이상, 정상 리뷰에서 많이 쓰이는 단어

```python
import matplotlib
from nltk import tokenize
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance
import seaborn as sns

from konlpy.tag import Okt
from collections import Counter
from wordcloud import WordCloud
```

이상, 정상 리뷰 많이 쓰이는 단어 워드 클라우드

```python
class word_count:
    def __init__(self):
        self.main()
        
    def main(self):

        f = open("C:/Users/User/OneDrive - 공주대학교/바탕 화면/review/after_negative(1).txt", 'r', encoding='utf-8')
        lines = f.readlines()
        text = ''

        for line in lines:
            line = line.strip()  # 줄 끝의 줄 바꿈 문자를 제거한다.
            text += line
        f.close()


        okt = Okt()
        okt_tokens = okt.morphs(text, stem=True)


        frequent = Counter(okt_tokens)
        print('이상 리뷰에서 많이 쓰이는 단어\n')
        print(frequent)
        
        self.visualize(frequent)  
        
        f = open('C:/Users/User/OneDrive - 공주대학교/바탕 화면/review/after_positive(0).txt',mode='r', encoding='utf-8' )

        lines = f.readlines()
        text = ''

        for line in lines:
            line = line.strip()  # 줄 끝의 줄 바꿈 문자를 제거한다.
            text += line
        f.close()


        okt = Okt()
        okt_tokens = okt.morphs(text, stem=True)


        frequent = Counter(okt_tokens)
        print('정상 리뷰에서 많이 쓰이는 단어\n')
        print(frequent)
        self.visualize(frequent)
        
    def visualize(self,list):
        wc = WordCloud(font_path='C:/Users/User/OneDrive - 공주대학교/바탕 화면/08 서울 남산체 B.ttf',background_color="white", width=1000, height=1000,max_words=100,max_font_size=300)
        wc.generate_from_frequencies(dict(list))
        plt.figure(figsize=(10,8))
        plt.axis('off')
        plt.imshow(wc)
        plt.show()

if __name__=="__main__":
    wc=word_count()
        
```
